```python
import matplotlib.pylab as plt
import numpy as np
```

```python
import csv
with open('elipse-40419.csv', 'rt') as f:
    reader = csv.reader(f)
    data = list(reader)
    for row in reader:
        print(row)
```

```python
tiempo = []
Xm = []
Ym = []
Ang = []
for i in data[1:]:
    tiempo.append(float(i[0]))
    Xm.append(float(i[2]))
    Ym.append(float(i[3]))
    Ang.append(float(i[6]))
```

```python
plt.figure(figsize=(10,10))
r = [tiempo,Xm,Ym,Ang]
for i in r[1:]:
    plt.subplot(3,1,r.index(i))
    plt.plot(r[0],i)
```

```python
with open('cilindroprueba5.csv', 'rt') as f:
    reader = csv.reader(f)
    data2 = list(reader)
    for row in reader:
        print(row)
```
